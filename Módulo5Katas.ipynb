Ejercicio 1 Utilizar Operadores Aritmeticos

```python
# Crear variables para almacenar las dos distancias
distanciaSolTierra = 149597870 #km  
distanciaSolJupiter = 778547200 #km
distanciakm =  distanciaSolJupiter - distanciaSolTierra
distanciami = distanciakm * 0.621371

print("Calcular distancia entre la tierra y jupiter\n")
print("La distancia en km entre la tierra y jupiter es de "+str(distanciakm))
print("La distancia en mi entre la tierra y jupiter es de "+str(distanciami))
```
