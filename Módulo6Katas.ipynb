Ejercicio 1 Crear y usar listas de python

```python
# Creamos la lista planets y la mostramos

planets = ['Mercurio', 'Venus', 'Tierra', 'Marte', 'Jupiter', 'Saturno', 'Urano', 'Neptuno']

#Mostrar números de planetas
number_planets = len(planets)
print("El número de planetas es",number_planets)

# Agregamos a plutón y mostramos el último elemento
planets.append('Pluton')
number_planets = len(planets)
print("\nEl número de planetas es",number_planets)
print('El ultimo planeta agregado es', planets[-1])
```

Ejercicio 2 Trabajando con datos de ua lista

```python
# Lista de planetas
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Neptune']

# Solicitamos el nombre de un planeta *Pista:  input()*
usuario_planeta = input("Introduce el nombre de un planeta del sistema solar usa la primer letra en mayuscula")

# Busca el planeta en la lista (index)
planeta_index = planets.index(usuario_planeta)
print(f'\nEl planeta agregado {usuario_planeta} es el ',planeta_index+1 , 'mas cercano al sol')

# Muestra los planetas más cercanos al sol
print(f'\nLos planetas mas cercanos al sol antes de {usuario_planeta} son: ')
planetasCercanosSol = planets[0:planeta_index]
print(planetasCercanosSol)

# Muestra los planetas más Lejanos al sol
print(f'\nLos planetas mas lejanos al sol despues de {usuario_planeta} son: ')
planetasLejanosSol = planets[planeta_index: ]
print(planetasLejanosSol)
```
