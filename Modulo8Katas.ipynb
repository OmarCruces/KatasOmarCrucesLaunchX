Ejercicio 1: Crear y modificar un diccionario

```python
#Ejercicio 1: Creación de un diccionario de Python

# Crea un diccionario llamado planet con los datos propuestos

planet = {
    'name': 'Mars',
    'moons': 2
}

# Muestra el nombre del planeta y el número de lunas que tiene.
print(planet['name'],'tiene',planet['moons'], 'moons')

# Agrega la clave circunferencia con los datos proporcionados previamente
planet['circunferencia (km)'] = { 
    'polar': 6752,
    'equatorial': 6792
}

# Imprime el nombre del planeta con su circunferencia polar.
print(planet['name'],'tiene',planet['circunferencia (km)']['polar'], 'km de circunferencia')
#print(f'{planet["name"]} tiene {planet["circunferencia (km)"]["polar"]}' de circunferencia)
```

Ejercicio 2: Programación dinámica con diccionarios

```python
# Planets and moons

planet_moons = {
    'mercury': 0,
    'venus': 0,
    'earth': 1,
    'mars': 2,
    'jupiter': 79,
    'saturn': 82,
    'uranus': 27,
    'neptune': 14,
    'pluto': 5,
    'haumea': 2,
    'makemake': 1,
    'eris': 1
}

# Añade el código para determinar el número de lunas.
#Lista de lunas
moons = planet_moons.values()

#Numero de planetas
planets = len(planet_moons.keys())


# Agrega el código para contar el número de lunas. 

total_moons = 0
for moon in moons:
    total_moons = total_moons + moon

#Promedio

promedio = total_moons / planets

print(promedio)
```
