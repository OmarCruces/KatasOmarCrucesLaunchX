```python
# Función con un informe preciso de la misión. 
# Considera hora de prelanzamiento, tiempo de vuelo, destino, tanque externo y tanque interno
def informe(hora_desplazo, tiempo_vuelo, destino, tanque_externo, tanque_interno):
    return f"""
    Misión a: {destino}
    Total tiempo de viaje: {hora_desplazo + tiempo_vuelo} minutos
    Total combustible: {tanque_externo + tanque_interno} galones
    """

print(informe(14, 51, "Luna", 200000, 300000))

```

```python
# Escribe tu nueva función de reporte considerando lo anterior
def informe(destino,*minutes,**fuel_reservoirs):
    return f"""
    Misión a: {destino}
    Total tiempo de viaje: {sum(minutes)} minutos
    Total combustible: {sum(fuel_reservoirs.values())} galones
    """

print(informe("Luna", 10, 25, main=300000, external=200000))
```

```python
# Escribe tu nueva función
def mision_reporte(destino, *minutes, **fuel_reservoirs):
    main_report = f"""
    Misión a: {destino}
    Total de tiempo de viaje: {sum(minutes)} minutos
    Total combustible: {sum(fuel_reservoirs.values())}
    """
    for tank_name, gallons in fuel_reservoirs.items():
        main_report += f"{tank_name} tank --> {gallons} gallones sobrantes\n"
    return main_report

print(mision_reporte("Luna", 8, 11, 55, main=300000, external=200000))
```
