Ejercicio 1: Uso de ciclos `while` en Python

```python
#Ejercicio 1: Creaci√≥n de un bucle while

#Definir las variable
new_planet = ''
planets = []

# Ciclo while
while new_planet.lower() != 'listo':
    if new_planet:
        planets.append(new_planet)
    new_planet = input('Entroduce el planeta, o listo ')

#Ejercicio 2: Creacion de un bucle for
    if new_planet == 'listo':
     print("Los planetas introducidos son\n")   
     for planet in planets:
       print(planet)
```
