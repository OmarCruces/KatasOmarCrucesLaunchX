```python
#Texto a trabajar
text = """Interesting facts about the Moon. The Moon is Earth's only satellite. There are several interesting facts about the Moon and how it affects life here on Earth. 
On average, the Moon moves 4cm away from the Earth every year. This yearly drift is not significant enough to cause immediate effects on Earth. The highest daylight temperature of the Moon is 127 C."""

#Dividir texto
texto_div = text.split('. ')
print(texto_div)

#Definir palabras
palabras = {"average", "temperature","distance"}

#Buscar palabras clave
for iteración in  texto_div:
   for word in palabras:
       if word in iteración:
           print(iteración)
           break
       
```

Cambiar C a Celsius

```python
#Texto a trabajar
text = """Interesting facts about the Moon. The Moon is Earth's only satellite. There are several interesting facts about the Moon and how it affects life here on Earth. 
On average, the Moon moves 4cm away from the Earth every year. This yearly drift is not significant enough to cause immediate effects on Earth. The highest daylight temperature of the Moon is 127 C."""

#Dividir texto
texto_div = text.split('. ')
print(texto_div)

#Definir palabras
palabras = {"average", "temperature","distance"}

#Buscar palabras clave
for iteración in  texto_div:
   for word in palabras:
       if word in iteración:
           print(iteración.replace('C', 'Celsius'))
           break
```

```python
# Datos con los que vas a trabajar
name = "Luna"
gravity = 0.00162 # in kms
planet = "Tierra"

# Creamos el título
titulo = "Datos de la gravedad en la {0} a diferencia de la  {1}".format(name,planet)
##titulo2 = f'Datos de la gravedad en la {name} a diferencia de la {planet}'
print(titulo)
##print(titulo2)

# Creamos la plantilla
hechos = f"La gravedad en la {name} es de {gravity*1000} m/s2 a diferencia de la {planet}"
print(hechos,'\n')

#Union de ambas cadenas
union = f"""{titulo},
-----------------------------
Nombre del planeta: {planet}
{hechos}
"""
print(union)
```
